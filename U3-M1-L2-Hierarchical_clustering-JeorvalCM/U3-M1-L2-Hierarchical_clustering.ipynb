# Assignment: Hierarchical Clustering

```python
# Imports
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster, cophenet
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

## Part 1 (10 pts)

```python
from sklearn.datasets import load_wine

data = load_wine()
X = data['data']
y = data['target']
```

This data is the result of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines.

1. Scale the data (clustering is sensitive to scale)
2. Find the distance between pairs of all 13-dimensional vectors. Use `pdist`.
3. Use linkage to create the clusters, try different linkage methods.
4. Visualize your clusters using `dendrogram`.
5. Get the top 3 clusters using `fcluster`.
6. Compare your top 3 clusters to the actual data.

```python
Y = list(map(lambda x: data['target_names'][x], y))
```

```python
### 1
scaler = StandardScaler()

X = scaler.fit_transform(X)

### 2
d_m = pdist(X)

### 3 'single' method
link_single = linkage(d_m)

### 4
plt.figure(figsize = (20, 12))
dendrogram(link_single, labels = Y, leaf_font_size = 5)
plt.show()
```

```python
### 3 'comple' methodc
link_complete = linkage(d_m, method = 'complete')

### 4
plt.figure(figsize = (20, 12))
dendrogram(link_complete, labels = Y, leaf_font_size = 5)
plt.show()
```

```python
### 3 'comple' methodc
link_average = linkage(d_m, method = 'average')

### 4
plt.figure(figsize = (20, 12))
dendrogram(link_average, labels = Y, leaf_font_size = 5)
plt.show()
```

```python
### 3 'comple' methodc
link_weighted = linkage(d_m, method = 'weighted')

### 4
plt.figure(figsize = (20, 12))
dendrogram(link_weighted, labels = Y, leaf_font_size = 5)
plt.show()
```

```python
### 3 'comple' methodc
link_ward = linkage(d_m, method = 'ward')

### 4
plt.figure(figsize = (20, 12))
dendrogram(link_ward, labels = Y, leaf_font_size = 5)
plt.show()
```

```python
# YOUR CODE HERE
### 5
clusters_labels_single = fcluster(link_single, 3, criterion = 'maxclust')

clusters_labels_complete = fcluster(link_complete, 3, criterion = 'maxclust')

clusters_labels_average = fcluster(link_average, 3, criterion = 'maxclust')

clusters_labels_weighted = fcluster(link_weighted, 3, criterion = 'maxclust')

clusters_labels_ward = fcluster(link_ward, 3, criterion = 'maxclust')
```

```python
# Build table with label count
# each row is a cluster
dict_to_pd = {key:0 for key in data['target_names']}

dict_clusters_single = {key: dict_to_pd.copy() for key in range(1, 4)}

dict_clusters_complete = {key: dict_to_pd.copy() for key in range(1, 4)}

dict_clusters_average = {key: dict_to_pd.copy() for key in range(1, 4)}

dict_clusters_weighted = {key: dict_to_pd.copy() for key in range(1, 4)}

dict_clusters_ward = {key: dict_to_pd.copy() for key in range(1, 4)}

for idx, val in enumerate(Y):
    
    dict_clusters_single[clusters_labels_single[idx]][val] += 1
    
    dict_clusters_complete[clusters_labels_complete[idx]][val] += 1
    
    dict_clusters_average[clusters_labels_average[idx]][val] += 1
    
    dict_clusters_weighted[clusters_labels_weighted[idx]][val] += 1
    
    dict_clusters_ward[clusters_labels_ward[idx]][val] += 1

single = pd.DataFrame(dict_clusters_single).T

complete = pd.DataFrame(dict_clusters_complete).T

average = pd.DataFrame(dict_clusters_average).T

weighted = pd.DataFrame(dict_clusters_weighted).T

ward = pd.DataFrame(dict_clusters_ward).T
```
